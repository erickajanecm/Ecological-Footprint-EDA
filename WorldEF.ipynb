<h2>World - Biocapacity vs Ecological Footprint (gha per person) 



```python
#Importing necessary libraries
import pandas as pd
```

```python
# Loading the data
World_EF = pd.read_csv('./Ecological Footprint vs Biocapacity (gha per person).csv')
World_EF.head()
```

```python
# Shape of the dataframe
World_EF.shape
```

```python
# Columns of the dataframe
World_EF.columns
```

There are columns in this dataframe that are not necessary for our analysis so we have to drop them.

```python
# Information about the dataframe
World_EF.info()
```

<h4> Transformation

```python
#Dropping the unnecessary columns
cols_to_drop = ['Country Name', 'Short Name', 'Data Quality Score', 'isoa2']
World_EF.drop(cols_to_drop, axis=1, inplace=True)

World_EF
```

```python
# Renaming the column 'year' - (capitalization)
World_EF.columns = World_EF.columns.str.replace('year', 'Year')

World_EF
```

<h4> Separating the 'BiocapPerCap' and 'EFConsPerCap' in the 'Record' column

```python
# EFConsPerCap
EF_DF = World_EF.loc[World_EF['Record'] == 'EFConsPerCap']

EF_DF.head()
```

```python
# BiocapPerCap
Biocapacity_DF = World_EF.loc[World_EF['Record'] == 'BiocapPerCap']
Biocapacity_DF.head()
```

<h4>Visualze the status of biocapacity and ecological footprint throughout the years

```python

#Importing libraries
import matplotlib.pyplot as plt
import numpy as np

x = Biocapacity_DF['Total'] 
year = np.arange(1961, 2019)
y = EF_DF['Total']
fig = plt.figure(figsize=(15, 7))
plt.plot(year, x, color='green')
plt.plot(year, y, color='red')

plt.title('World\'s Decline in Biocapacity and Increase in Ecological Footprint from 1961-2018')
plt.xlabel('Year')
plt.ylabel('gha per person')
plt.xticks(ticks=year, rotation=70)
plt.grid(True)
plt.grid(color='red',  alpha=0.2, linewidth=2)
plt.legend(labels=['Biocapcity', 'Ecological Footprint'])
```
