Many columns does not provide useful insights so we will drop those columns

```python
#Deleting unnecessary records
cols_to_drop = ['m49_code', 'region', 'cpc_code', 'sample_size', 'method_data_collection','reference', 'url', 'notes' ]
FoodWaste_DF.drop(columns=cols_to_drop, inplace=True)
```

```python
FoodWaste_DF.head(10)
```

```python
#Converting Columns to uppercase
FoodWaste_DF.columns = [x.upper() for x in FoodWaste_DF.columns]
FoodWaste_DF.head()
```

```python
FoodWaste_DF.info()
```

```python
#Looking for different countries
FoodWaste_DF['COUNTRY'].unique()
```

```python
#South East Asian Countries records
SEA_Countries = ['Indonesia', 'Timor Leste', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Cmabodia', 'Laos', 'Thailand', 'Vietname']
SEA_Countries_FoodWaste = FoodWaste_DF.loc[FoodWaste_DF['COUNTRY'].isin(SEA_Countries)]
SEA_Countries_FoodWaste.head()
```

```python
SEA_Countries_FoodWaste.isna().sum()
```
