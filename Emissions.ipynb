```python
#CREATING A DATAFRAME FOR 'PHILIPPINES'

PH_Emission_DF = Emission_DF[Emission_DF['Nation'] == 'PHILIPPINES']
PH_Emission_DF
```

```python
year = PH_Emission_DF['Year'][-18:].tolist()
y_values = np.arange(0, 40000)
solidFuel = PH_Emission_DF['Emissions from solid fuel consumption'][-18:]
liquidFuel = PH_Emission_DF['Emissions from liquid fuel consumption'][-18:]
gasFuel = PH_Emission_DF['Emissions from gas fuel consumption'][-18:]
fig = plt.figure(figsize=(15, 8))
plt.plot(year, solidFuel, color='red')
plt.plot(year, liquidFuel, color='green')
plt.plot(year, gasFuel, color='green')

plt.title('Philippines Emission from solid fuel consumption 1961-2018')
plt.xlabel('Year')
plt.ylabel('Metric tons of Carbon')
plt.xticks(ticks=year, rotation=70)
plt.yticks(ticks=y_values)
plt.legend(labels=['Solid Fuel Consumption', 'Liquid Fuel Consumption', 'Gas Fuel Consumption'])
```
